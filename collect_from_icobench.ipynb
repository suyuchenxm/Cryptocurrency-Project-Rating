```python
def get_financial_data(soup):
    finSess=soup.find('div',{'class':'tab_content','id':'financial'})
    tb=dict()
    for row in finSess.find_all('div',{'class':'row'}):
        try:
            tb[row.find('div',{'class':'label'}).get_text().strip()]=row.find('div',{'class':'value'}).get_text().strip()
        except:
            pass
    return tb
```

```python
def get_team_data(project_name,soup):
    try:
        teamSess=soup.find('div',{'class':'tab_content','id':'team'}).find_all('div',{'class':'col_3 iss'})
        team_data=[]
        for member in teamSess:
            try:
                name=member.find('h3').get_text().strip()
                title=member.find('h4').get_text().strip()
                linkedin=member.find('a',{'class':'linkedin'}).get('href')
                team_data.append([project_name,name,title,linkedin])
            except:
                pass
        return team_data
    except:
        print(f'get project {project_name} team data error')
        return [project_name,None,None,None]
```

```python
finData={}
teamData=[]
for idx in urls.index:
    url=urls.loc[idx,'url']
    try:
        time.sleep(random.randint(1,10))
        soup=bs(requests.get('https://icobench.com'+url).content,'lxml')
        project_name=soup.find('div',{'class':'name'}).find('h1').get_text().strip()
        urls.loc[idx,'Project name']=project_name
        finData[project_name]=get_financial_data(soup)
        teamData.extend(get_team_data(project_name,soup))
        print(f'finish scraping {url}')
    except:
        print('getting error +1')
        pass
  
```

```python
def clean_name(item):
    name=re.sub(' ','',item)
    name=name.lower()
    return name

def clean_ICO_price(item):
    try:
        if 'USD' in item:
            price=re.sub('USD','',item)
            if '=' in price:
                price=re.sub('^[a-zA-Z0-9\s]+=','',price).strip()
        else:
            price=''
        return price
    except:
        pass
    
```

```python
def drill(list_):
    try:
        a=next(item for item in list_ if item is not 'NAN')
        return next(item for item in list_ if item is not 'NAN')
    except:
        pass
```
